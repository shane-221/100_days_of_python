```python
import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format
dt= pd.read_csv("Dataset_salaries_by_college_major.csv")
dt.head()
dt.shape
dt.columns
dt.isna()
dt.tail()
clean_dt= dt.dropna()
clean_dt.tail()
clean_dt["Starting Median Salary"]
clean_dt["Starting Median Salary"].max()
value = clean_dt["Starting Median Salary"].idxmax()


max_starting_salary = clean_dt["Starting Median Salary"][value]
print(max_starting_salary)

# entire row to get the location value
clean_dt.loc[value]
```

```python
# Questions
## Now that we've found the major with the highest starting salary, can you write the code to find the following:

 ### What college major has the highest mid-career salary? How much do graduates with this major earn? (Mid-career is defined as having 10+ years of experience).

### Which college major has the lowest starting salary and how much do graduates earn after university?

### Which college major has the lowest mid-career salary and how much can people expect to earn with this degree?
```

```python
### Question 1
max_mid_career_salary = clean_dt["Mid-Career Median Salary"].max()
row_value = clean_dt[clean_dt['Mid-Career Median Salary']== max_mid_career_salary]
row_value["Undergraduate Major"]


### Question 2
lowest_starting_salary= clean_dt['Starting Median Salary'].min()
row_value =clean_dt[clean_dt["Starting Median Salary"]==lowest_starting_salary]

row_value["Undergraduate Major"]

###  Question 3:
min_mid_career_salary = clean_dt["Mid-Career Median Salary"].min()
row_value = clean_dt[clean_dt['Mid-Career Median Salary']== min_mid_career_salary]
major = row_value["Undergraduate Major"]
print(f" Lowest Paying undergraduate Major:{major}")
print(f" Lowest Paying undergraduate Major salary:{min_mid_career_salary}")
```

```python
### Calculate the difference between differnt values. E.g. How would we calculate the difference between the earnings of the 10th a 90th percentile?


spread_col = clean_dt["Mid-Career 90th Percentile Salary"]- clean_dt["Mid-Career 10th Percentile Salary"]
clean_dt.insert(1, "Spread", spread_col)
print(clean_dt.head())
```

```python
# Sorting the Spread column
### Low risk sectors are the ones where there is low spread
low_risk =clean_dt.sort_values("Spread")
low_risk[["Undergraduate Major" , "Spread"]]
```
